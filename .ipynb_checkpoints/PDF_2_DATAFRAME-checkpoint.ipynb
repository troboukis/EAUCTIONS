```python
import tika
import requests
from tika import parser
import pandas as pd
import re
import os
pd.set_option('display.max_columns',1000)
pd.set_option('display.width', 1000)
# Start running the tika service
tika.initVM()
```

```python
import glob

filenames = glob.glob('/Users/viceproduction/Google Drive/Documentation/*')

fp = pd.DataFrame({'filepath': filenames})
fp.head()
```

```python
def process_file(filepath):
    results=parser.from_file(filepath)
    metadata=results.get('metadata')

    data = pd.Series({
        'status': results['status'],
        'content': results.get('content'),
        'filename': os.path.basename(filepath),
        
    })
    if metadata is not None:
        
        if 'Content-Type' in metadata:
            data['Content-Type'] = metadata['Content-Type']
            
        if 'Creation-Date' in metadata:
            data['Creation-Date'] = metadata['Creation-Date']

        if 'xmpTPg:NPages' in metadata:
            data['Pages'] = metadata['xmpTPg:NPages']
        if 'xmpMM:DocumentID' in metadata:
            data['Document-id'] = metadata['xmpMM:DocumentID']

        
    return data
```

```python
df=fp.filepath.apply(process_file)
```

```python
df.to_csv('data_from_pdfs.csv',index=False)
```

```python
dpdf=pd.read_csv('data_from_pdfs.csv')
```

```python
def clean_series (series):
    return str(series).strip().replace('\n',' ')
```

```python
dpdf['content']=dpdf['content'].apply(lambda x: clean_series(x))
```

```python
dpdf.to_csv('data_from_pdfs_cleaned_content.csv',index=False)
```
