```python
# Function to clean the unique code:

def code_cleaning(dataframe):
    return re.split('[-]',str(dataframe))
```

```python
df['code']=df['Unique_Code'].apply(lambda x:code_cleaning(x)[0]) 
```

```python
# Walk through all files in the directory that contains the files to copy
import time
movdir=r"/Users/viceproduction/Google Drive/Documentation/test/"
basedir=r"/Users/viceproduction/Desktop/test/"

a=0
for i, fnm in enumerate (df['PDF']):
    for root, dirs, files in os.walk(movdir):
        for filename in files:
            base, extension = os.path.splitext(filename)
        
            if filename==fnm:
                a+=1 
                old_name = os.path.join( os.path.abspath(root), filename )
                new_name = os.path.join(basedir, df['Unique_Code'][i]+extension)

                shutil.copy(old_name, new_name)
                                
                print('Copied {} files'.format(a))
                clear_output(wait=True)
```
